## Need Generalization
We need to learn in a way whose complexity doesn't scale $\Longrightarrow$ find $\theta$ s.t. $f(\cdot, \theta) = V^{*}(\cdot)$

This is what the supervised learning is good at.

<img src="images/lecture_0/generalization.png" alt="Image Description" width="600" height="600">

Difference between supervised learning and bandit problem is the label(ofcause data is generated by itself is also a difference). In supervised lenaring, the learner knows which action is the best, on the other hand, in bandit problem, learner only get feedback how good the decision you made.
